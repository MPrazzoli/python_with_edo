# k-means

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
clust = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 12345)
```

```python
plt.scatter(clust[0][:,0], clust[0][:,1], c = clust[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(clust[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
cl1 = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 12345)
    kmeans.fit(clust[0])
    cl1.append(kmeans.inertia_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 10), cl1)
plt.title('Elbow method')
plt.xlabel('Numero di cluster')
plt.ylabel('Var')
```

# Secondo esempio su dataset iris

```python
df = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
```

```python
df.tail(20)
df['Iris-setosa'].value_counts()
```

```python
cl = df.iloc[:, 4]
```

```python
df2 = df.iloc[:, 0:4]
```

```python
print(df2.head(4))
```

```python
kmeans = KMeans(n_clusters = 3)
```

```python
kmeans.fit(df2)
```

```python
kmeans.labels_
```

```python
cl1 = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 12345)
    kmeans.fit(df2)
    cl1.append(kmeans.inertia_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 10), cl1)
plt.title('Elbow method')
plt.xlabel('Numero di cluster')
plt.ylabel('Var')
```

```python
cl1 = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 0}
df.iloc[:, 4] = df.iloc[:, 4].map(cl1)
```

```python
cl_n = df.iloc[:, 4]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(kmeans.labels_, cl_n))
print(classification_report(kmeans.labels_, cl_n))
```
